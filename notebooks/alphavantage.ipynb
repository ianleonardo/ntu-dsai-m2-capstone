```python
import os
import requests
from dotenv import load_dotenv
import json

# Load environment variables to get VANTAGE_API_KEY
load_dotenv()
api_key = os.getenv('VANTAGE_API_KEY')

ticker = 'IBM'
```

```python
url = f'https://www.alphavantage.co/query?function=RSI&symbol={ticker}&interval=weekly&time_period=10&series_type=open&apikey={api_key}'
r = requests.get(url)
data = r.json()

print(json.dumps(data, indent=4))
```

```python
# Income statement
url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}'
r = requests.get(url)
data = r.json()

print(json.dumps(data, indent=4))
```

```python
# Earning history
# This API returns the annual and quarterly earnings (EPS) for the company of interest. 

url = f'https://www.alphavantage.co/query?function=EARNINGS&symbol={ticker}&apikey={api_key}'
r = requests.get(url)
data = r.json()

print(json.dumps(data, indent=4))
```
